# USA States Map With CONUS Albers Projection

```kotlin
%useLatestDescriptors
%use kandy-geo
```

```kotlin
val usaStates = GeoDataFrame.readGeoJson("https://raw.githubusercontent.com/AndreiKingsley/datasets/refs/heads/main/USA.json")
```

```kotlin
val usaAlbers = usaStates
    .modify { filter { name !in listOf("Alaska", "Hawaii", "Puerto Rico") } }
    .applyCrs(CRS.decode("EPSG:5070", true))
```

```kotlin
usaAlbers.plot {
    geoPolygon()
    layout.title = "US With CONUS Albers Projection"
}
```
