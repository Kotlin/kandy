# German States Map With Simple Settings (Shapefile)

```kotlin
%useLatestDescriptors
%use kandy-geo
```

```kotlin
val worldStates = GeoDataFrame.readShapefile("https://github.com/AndreiKingsley/datasets/raw/refs/heads/main/ne_10m_admin_1_states_provinces/ne_10m_admin_1_states_provinces.shp")
```

```kotlin
val germanStates = worldStates.modify {
    filter { admin == "Germany" }
}
```

```kotlin
germanStates.plot {
    geoMap {
        fillColor = Color.hex("#E0E8D3")

        borderLine {
            width = 0.35
            color = Color.hex("#567A56")
        }
    }
    layout {
        title = "Germany States"
        layout.style(Style.Void)
    }
}
```
