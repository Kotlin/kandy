## `Free` scales on faceted plot

```kotlin
%useLatestDescriptors
%use ggdsl(0.2.4-dev-5)
```

```kotlin
var mpg_df = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-kotlin/master/docs/examples/data/mpg2.csv")
mpg_df.head()
```

```kotlin
mpg_df.create { plot { 
    x(`engine horsepower`)
    y("engine displacement (cu. inches)"<Double>())
    point { 
        color(`origin of car`)
    }
    layout { 
        theme(Theme.Grey)
        size = 800 to 350
    }
} }
```

### Faceted plot

#### `facetGrid()` with `fixed` scales (the default)

Scales are constant across all panels.

```kotlin
mpg_df.create { plot { 
    x(`engine horsepower`)
    y("engine displacement (cu. inches)"<Double>())
    point { 
        color(`origin of car`)
    }
    layout { 
        theme(Theme.Grey)
        size = 800 to 500
    }
    facetGridY(`origin of car`)
} }
```

#### `facetGrid()` with `free` Y-scales

```kotlin
mpg_df.create { plot { 
    x(`engine horsepower`)
    y("engine displacement (cu. inches)"<Double>())
    point { 
        color(`origin of car`)
    }
    layout { 
        theme(Theme.Grey)
        size = 800 to 500
    }
    facetGridY(`origin of car`, scalesSharing = ScalesSharing.FREE_Y)
} }
```

#### `facetWrap()` with `fixed` scales (the default)

Scales are constant across all panels.

```kotlin
mpg_df.create { plot { 
    x(`engine horsepower`)
    y("engine displacement (cu. inches)"<Double>())
    point { 
        color(`origin of car`)
    }
    layout { 
        theme(Theme.Grey)
        size = 800 to 500
    }
    facetWrap() {
        facet(`number of cylinders`, order = OrderDirection.ASCENDING)
    }
} }
```

#### `facetWrap()` with `free` scales along both axis

```kotlin
mpg_df.create { plot { 
    x(`engine horsepower`)
    y("engine displacement (cu. inches)"<Double>())
    point { 
        color(`origin of car`)
    }
    layout { 
        theme(Theme.Grey)
        size = 800 to 500
    }
    facetWrap(scalesSharing = ScalesSharing.FREE) {
        facet(`number of cylinders`, order = OrderDirection.ASCENDING)
    }
} }
```
